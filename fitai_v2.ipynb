```python
import random
import pandas as pd

def generate_sample_dataset(num_samples):

    def generate_age():
        return random.randint(18, 65)

    def generate_gender():
        return random.choice(['M', 'F'])

    def generate_height():
        return round(random.uniform(150, 200), 2)

    def generate_weight():
        return round(random.uniform(50, 100), 2)

    def calculate_bmi(weight, height):
        height_m = height / 100
        return round(weight / (height_m ** 2), 2)

    def generate_fitness_level():
        return random.choice(['Beginner', 'Intermediate', 'Advanced'])

    def generate_medical_conditions():
        conditions = ['None', 'Hypertension', 'Diabetes', 'Asthma', 'Heart Disease']
        return random.choice(conditions)

    def generate_injuries():
        injuries = ['None', 'Knee Pain', 'Back Pain', 'Shoulder Pain', 'Ankle Sprain']
        return random.choice(injuries)

    def generate_fitness_goals():
        goals = ['Weight Loss', 'Muscle Gain', 'Endurance', 'Flexibility']
        return random.choice(goals)

    def generate_preferred_exercise_types():
        types = ['Cardio', 'Strength', 'Yoga', 'Pilates', 'HIIT']
        return random.sample(types, random.randint(1, 3))

    def generate_exercise_frequency():
        return random.randint(1, 7)

    def generate_exercise_duration():
        return random.randint(30, 90)

    def generate_preferred_time_of_day():
        return random.choice(['Morning', 'Afternoon', 'Evening'])

    def generate_preferred_environment():
        return random.choice(['Gym', 'Home', 'Outdoors'])

    def generate_activity_level():
        return random.choice(['Sedentary', 'Lightly Active', 'Active'])

    def generate_dietary_preferences():
        preferences = ['None', 'Vegetarian', 'Vegan', 'Keto', 'Paleo']
        return random.choice(preferences)

    def generate_sleep_patterns():
        return random.uniform(5, 9)

    def generate_motivation_level():
        return random.choice(['Low', 'Moderate', 'High'])

    def generate_stress_level():
        return random.choice(['Low', 'Moderate', 'High'])

    def generate_exercise_history():
        return random.choice(['None', 'Regular', 'Occasional'])

    def generate_favorite_exercises():
        exercises = ['Running', 'Cycling', 'Yoga', 'Swimming', 'Pilates', 'HIIT', 'Jogging', 'Marching on the Spot', 'Arm Circles', 
                     'Seated Rowing', 'Modified Wall Push-Ups', 'Side Leg Lifts', 'Water Bottle Bicep Curls', 'Brisk Walking', 
                     'Wall Push-Ups', 'Chair Squats', 'Step-ups', 'Modified Plank', 'Air Squats', 'Push-Ups', 'Walking Lunges', 
                     'Dumbbell Rows', 'Plank', 'Jumping Jacks', 'Bodyweight Squats', 'Lunges', 'Glute Bridges', 'Tricep Dips on Chair', 
                     'Supermans', "Incline Push-Ups", 'Bird Dogs', "Burpees", 'Dips', 'Elevated Pike Pushups', 'Bulgarian Split Squats', 
                     'Assisted Pistol Squats', 'Sliding Leg Curl', 'Seated Leg Lifts', 'Reverse Lunges']
        return random.sample(exercises, random.randint(1, 3))

    def generate_disliked_exercises():
        exercises = ['Running', 'Cycling', 'Yoga', 'Swimming', 'Pilates', 'HIIT', 'Jogging', 'Marching on the Spot', 'Arm Circles', 
                     'Seated Rowing', 'Modified Wall Push-Ups', 'Side Leg Lifts', 'Water Bottle Bicep Curls', 'Brisk Walking', 
                     'Wall Push-Ups', 'Chair Squats', 'Step-ups', 'Modified Plank', 'Air Squats', 'Push-Ups', 'Walking Lunges', 
                     'Dumbbell Rows', 'Plank', 'Jumping Jacks', 'Bodyweight Squats', 'Lunges', 'Glute Bridges', 'Tricep Dips on Chair', 
                     'Supermans', "Incline Push-Ups", 'Bird Dogs', "Burpees", 'Dips', 'Elevated Pike Pushups', 'Bulgarian Split Squats', 
                     'Assisted Pistol Squats', 'Sliding Leg Curl', 'Seated Leg Lifts', 'Reverse Lunges']
        return random.sample(exercises, random.randint(1, 3))

    def generate_available_equipment():
        equipment = ['Dumbbells', 'Treadmill', 'Resistance Bands', 'Kettlebells', 'None']
        return random.sample(equipment, random.randint(1, 3))

    def generate_weather_conditions():
        return random.choice(['Sunny', 'Rainy', 'Snowy', 'Windy'])

    def generate_gym_membership():
        return random.choice(['Yes', 'No'])

    def generate_support_system():
        return random.choice(['Yes', 'No'])

    def generate_group_classes_preference():
        return random.choice(['Yes', 'No'])

    def generate_recommended_exercises(favorite_exercises, disliked_exercises):
        all_exercises = set(['Running', 'Cycling', 'Yoga', 'Swimming', 'Pilates', 'HIIT', 'Jogging', 'Marching on the Spot', 'Arm Circles', 
                             'Seated Rowing', 'Modified Wall Push-Ups', 'Side Leg Lifts', 'Water Bottle Bicep Curls', 'Brisk Walking', 
                             'Wall Push-Ups', 'Chair Squats', 'Step-ups', 'Modified Plank', 'Air Squats', 'Push-Ups', 'Walking Lunges', 
                             'Dumbbell Rows', 'Plank', 'Jumping Jacks', 'Bodyweight Squats', 'Lunges', 'Glute Bridges', 'Tricep Dips on Chair', 
                             'Supermans', "Incline Push-Ups", 'Bird Dogs', "Burpees", 'Dips', 'Elevated Pike Pushups', 'Bulgarian Split Squats', 
                             'Assisted Pistol Squats', 'Sliding Leg Curl', 'Seated Leg Lifts', 'Reverse Lunges'])
        # Convert set to list before sampling
        available_exercises = list(all_exercises - set(disliked_exercises))
        recommended = set(random.sample(available_exercises, random.randint(12, 20)))
        recommended.update(favorite_exercises)
        return list(recommended)

    # Create a dictionary to hold the dataset
    data = {
        'Age': [],
        'Gender': [],
        'Height': [],
        'Weight': [],
        'BMI': [],
        'FitnessLevel': [],
        'MedicalConditions': [],
        'Injuries': [],
        'FitnessGoals': [],
        'PreferredExerciseTypes': [],
        'ExerciseFrequency': [],
        'ExerciseDuration': [],
        'PreferredTimeOfDay': [],
        'PreferredEnvironment': [],
        'ActivityLevel': [],
        'DietaryPreferences': [],
        'SleepPatterns': [],
        'MotivationLevel': [],
        'StressLevel': [],
        'ExerciseHistory': [],
        'FavoriteExercises': [],
        'DislikedExercises': [],
        'AvailableEquipment': [],
        'WeatherConditions': [],
        'GymMembership': [],
        'SupportSystem': [],
        'GroupClassesPreference': [],
        'RecommendedExercises': []
    }

    for i in range(num_samples):
        age = generate_age()
        height = generate_height()
        weight = generate_weight()
        bmi = calculate_bmi(weight, height)

        favorite_exercises = generate_favorite_exercises()
        disliked_exercises = generate_disliked_exercises()

        data['Age'].append(age)
        data['Gender'].append(generate_gender())
        data['Height'].append(height)
        data['Weight'].append(weight)
        data['BMI'].append(bmi)
        data['FitnessLevel'].append(generate_fitness_level())
        data['MedicalConditions'].append(generate_medical_conditions())
        data['Injuries'].append(generate_injuries())
        data['FitnessGoals'].append(generate_fitness_goals())
        data['PreferredExerciseTypes'].append(generate_preferred_exercise_types())
        data['ExerciseFrequency'].append(generate_exercise_frequency())
        data['ExerciseDuration'].append(generate_exercise_duration())
        data['PreferredTimeOfDay'].append(generate_preferred_time_of_day())
        data['PreferredEnvironment'].append(generate_preferred_environment())
        data['ActivityLevel'].append(generate_activity_level())
        data['DietaryPreferences'].append(generate_dietary_preferences())
        data['SleepPatterns'].append(generate_sleep_patterns())
        data['MotivationLevel'].append(generate_motivation_level())
        data['StressLevel'].append(generate_stress_level())
        data['ExerciseHistory'].append(generate_exercise_history())
        data['FavoriteExercises'].append(favorite_exercises)
        data['DislikedExercises'].append(disliked_exercises)
        data['AvailableEquipment'].append(generate_available_equipment())
        data['WeatherConditions'].append(generate_weather_conditions())
        data['GymMembership'].append(generate_gym_membership())
        data['SupportSystem'].append(generate_support_system())
        data['GroupClassesPreference'].append(generate_group_classes_preference())
        data['RecommendedExercises'].append(generate_recommended_exercises(favorite_exercises, disliked_exercises))

    # Convert the dictionary to a pandas DataFrame
    df = pd.DataFrame(data)
    
    return df

```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
import numpy as np

numerical_cols = features.select_dtypes(include=[np.number]).columns
categorical_cols = features.select_dtypes(include=[object]).columns

numetrans = Pipeline(
    steps=[
        ("impute", SimpleImputer(strategy="mean")),
        ("scale", StandardScaler())
    ]
)

ohe_cols = ['Gender', 'PreferredTimeOfDay', 'PreferredEnvironment', 'WeatherConditions', 'GymMembership', 'SupportSystem', 'GroupClassesPreference']
ordinal_cols = ['FitnessLevel', 'ActivityLevel', 'MotivationLevel', 'StressLevel', 'ExerciseHistory']

catetrans = Pipeline(
    steps=[
        ('impute', SimpleImputer(strategy='most_frequent')),
        ('ohe', OneHotEncoder(handle_unknown='ignore', sparse=False)),
        ('ordinal', OrdinalEncoder())
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numetrans, numerical_cols),
        ('cat', catetrans, categorical_cols)
    ]
)

pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('model', RandomForestClassifier(n_estimators=10, random_state=42))
    ]
)
```

```python
from sklearn.metrics import hamming_loss
y_pred = pipeline.predict(X_test)

#  calculate the average accuracy for all the classes
from sklearn.metrics import accuracy_score

def multioutput_accuracy_score(y_true, y_pred):
    return np.mean([accuracy_score(y_true.iloc[:, i], y_pred[:, i]) for i in range(y_true.shape[1])])

multioutput_accuracy_score(y_test, y_pred)

print(hamming_loss(y_test, y_pred))
print(multioutput_accuracy_score(y_test, y_pred))   
```
